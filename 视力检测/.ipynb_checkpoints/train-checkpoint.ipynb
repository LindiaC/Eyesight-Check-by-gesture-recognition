```python
from PIL import Image
import matplotlib.pyplot as plt
import cv2
model = cls(backbone='LeNet')
cap = cv2.VideoCapture(0)
ret, img = cap.read()
cap.release()
img =  cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img=cv2.flip(img, 1)
result = model.inference(image=img, show=True, checkpoint = checkpoint)
plt.imshow(img)
print(result)
model.print_result(result)
```

```python
import cv2

import BaseDeploy as bd


model_path = 'out_file/hand.onnx'
model = bd(model_path)


import cv2
import time



cap = cv2.VideoCapture(0)
end=time.time()+3

while True:
    ret, stream= cap.read()
    stream =  cv2.cvtColor(stream, cv2.COLOR_BGR2RGB)
    stream=cv2.flip(stream, 1)
    plt.imshow("stream")
    if time.time()>end:
        break

print("ok")
ret, img = cap.read()
cap.release()
result = model.inference(img)
result = model.print_result(result)

img =  cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img=cv2.flip(img, 1)
plt.imshow(img)
print(result)

 
```
